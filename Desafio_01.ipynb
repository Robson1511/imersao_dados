#Importando Base de Dados

```python

import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados.head()
```

## Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

```python
dados.query('tratamento == "com_controle"').value_counts()


## Acredito que o desbalanceamento se deve ao fato de a ideia principal da pesquisa é testar as drogas para analisar seu comportamento,
## tendo em vista a analise dos testes sem drogas à nível de comparação.
```

## Desafio 02: Plotar as 5 últimas linhas da tabela

```python
dados['tratamento'].tail()
```

## Desafio 03: Proporção das classes tratamento.

```python
dados['tratamento'].value_counts(normalize = True).round(2)
```

## Desafio 04: Quantas tipos de drogas foram investigados?

```python
dados['droga'].value_counts()

# Resporta: foram testados 3.289 tipos de drogas diferentes.
```
